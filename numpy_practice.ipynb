```python
#finding dimensions
a = np.array(10)
b = np.array([1,2,3,4,5,6,7,8])
c = np.array([[1,2,3],[4,5,6]])
d = np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
#finding dimensions with ndim function
print('a dimension :',a.ndim)
print('b dimension :',b.ndim)
print('c dimension :',c.ndim)
print('d dimension :',d.ndim)
#finding data type of a,b,c and d
print(a.dtype)
print(b.dtype)
print(c.dtype)
print(d.dtype)
#finding which type of class
print(type(a))
print(type(b))
print(type(c))
print(type(d))
```
